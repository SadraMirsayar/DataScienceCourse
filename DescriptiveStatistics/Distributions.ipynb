```python
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# rvs : random variable sample.
# size : number of random variable sample.
# loc : mean
# scale : std
data_normal = stats.norm.rvs(size=100000, loc=0, scale=1)
```

```python
print('data_normal: ', data_normal)

print('data_normal LENGTH: ', len(data_normal))

print('data_normal TYPE: ', type(data_normal))
```

```python
# https://seaborn.pydata.org/generated/seaborn.displot.html
# displot : distribution plot
# kde plot
sns.displot(data_normal, bins=100, kde=False)
```

```python
data_uniform = stats.uniform.rvs(size=100000, loc=0, scale=10)
```

```python
sns.displot(data_uniform, bins=100 ,kde=False)
```

```python
# p : win probability
# 0.2 is 20% win probability
data_bernoulli = stats.bernoulli.rvs(size=1000, p=0.2)
print(data_bernoulli)
```

```python
sns.displot(data_bernoulli, bins=100, kde=False)
```

```python
data_binomial = stats.binom.rvs(size=1000, n=100, p=0.2)
print(data_binomial)
```

```python
# data_binomial displot is DISCRETE!
sns.displot(data_binomial, bins=50, kde=False )
```

```python
import numpy as np
```

```python
# p : probability
p = 1 / 6
# n : number of Sub_Labs!
data_multinomial = np.random.multinomial(n=6, pvals=[p, p, p, p, p, p], size=1000)
print(data_multinomial)
```

```python
icu = 0.3
surgery = 0.2
el = 0.5
data_multinomial = np.random.multinomial(n=1000, pvals=[icu, surgery, el], size=365)
print(data_multinomial)
```

```python
# numbers must to be pre-define! (discrete)
data_poisson = np.random.poisson(lam=7, size=1000)
print(data_poisson)
```

```python
sns.displot(data_poisson, kde=False)
```

```python
data_exponential = np.random.exponential(scale=0.2, size=1000)
print(data_exponential)
```

```python
sns.displot(data_exponential, kde=False)
```
