```python
from scipy import stats
from math import sqrt
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_bar = 102
n = 100
sigma = 16
mu = 100
```

```python
z = (x_bar - mu) / (sigma / sqrt(n))
print(z)
```

```python
plt.style.available
```

```python
# Syntax: plt.style.use('style_name”)
plt.style.use('seaborn-v0_8')
plt.fill_between(x=np.arange(-4, z, 0.01), y1=stats.norm.pdf(np.arange(-4, z, 0.01)), facecolor='green')
plt.fill_between(x=np.arange(z, 4, 0.01), y1=stats.norm.pdf(np.arange(z, 4, 0.01)), facecolor='blue')
```

```python
stats.norm.cdf(z)
```

```python
p_value = 1 - stats.norm.cdf(z)
```
