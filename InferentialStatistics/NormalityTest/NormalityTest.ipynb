```python
from scipy.stats import shapiro
import numpy as np
```

```python
data = np.random.normal(loc=20, scale=5, size=1000)
print(data)
```

```python
# shapiro(data)
# Threshold is 5%

chi, p_value = shapiro(data)
print('chi:', chi)
print('p_value:', p_value)
```

```python
import scipy.stats as stats
import pylab
```

```python
# probability.Plot uses the concept of Cumulative distribution function
# probplot = probability plot
stats.probplot(data, dist='norm', plot=pylab)
```
